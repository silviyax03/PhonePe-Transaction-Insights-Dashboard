```python
import os
import json
import pandas as pd
import sqlite3
```

```python
path = "/content/pulse/data/aggregated/transaction/country/india/state"

records = []

for state in os.listdir(path):
    state_path = os.path.join(path, state)

    if os.path.isdir(state_path):
        for year in os.listdir(state_path):
            year_path = os.path.join(state_path, year)

            for file in os.listdir(year_path):
                if file.endswith(".json"):

                    with open(os.path.join(year_path, file), "r") as f:
                        data = json.load(f)

                    quarter = int(file.replace(".json", ""))

                    try:
                        transactions = data["data"]["transactionData"]

                        if transactions:
                            for item in transactions:
                                records.append({
                                    "State": state.replace("-", " ").title(),
                                    "Year": int(year),
                                    "Quarter": quarter,
                                    "Transaction_Type": item["name"],
                                    "Count": item["paymentInstruments"][0]["count"],
                                    "Amount": item["paymentInstruments"][0]["amount"]
                                })
                    except:
                        pass

df = pd.DataFrame(records)

df.head()
```

```python
conn = sqlite3.connect("phonepe.db")

df.to_sql(
    "aggregated_transaction",
    conn,
    if_exists="replace",
    index=False
)

print("Data inserted successfully")
```

```python
path = "/content/pulse/data/aggregated/user/country/india/state"

user_records = []

for state in os.listdir(path):
    state_path = os.path.join(path, state)

    if os.path.isdir(state_path):
        for year in os.listdir(state_path):
            year_path = os.path.join(state_path, year)

            for file in os.listdir(year_path):
                if file.endswith(".json"):

                    with open(os.path.join(year_path, file), "r") as f:
                        data = json.load(f)

                    quarter = int(file.replace(".json", ""))

                    try:
                        users = data["data"]["usersByDevice"]

                        if users:
                            for user in users:
                                user_records.append({
                                    "State": state.replace("-", " ").title(),
                                    "Year": int(year),
                                    "Quarter": quarter,
                                    "Brand": user["brand"],
                                    "Count": user["count"],
                                    "Percentage": user["percentage"]
                                })

                    except:
                        pass

df_user = pd.DataFrame(user_records)

df_user.head()
```

```python
import sqlite3

conn = sqlite3.connect("phonepe.db")

df_user.to_sql(
    "aggregated_user",
    conn,
    if_exists="replace",
    index=False
)

print("aggregated_user table created successfully!")
```

```python
path = "/content/pulse/data/aggregated/insurance/country/india/state"

insurance_records = []

for state in os.listdir(path):
    state_path = os.path.join(path, state)

    if os.path.isdir(state_path):

        for year in os.listdir(state_path):
            year_path = os.path.join(state_path, year)

            for file in os.listdir(year_path):

                if file.endswith(".json"):

                    quarter = int(file.replace(".json", ""))

                    with open(os.path.join(year_path, file), "r") as f:
                        data = json.load(f)

                    try:
                        insurance_data = data["data"]["transactionData"]

                        if insurance_data:

                            for item in insurance_data:

                                insurance_records.append({
                                    "state": state.replace("-", " ").title(),
                                    "year": int(year),
                                    "quarter": quarter,
                                    "insurance_type": item["name"],
                                    "insurance_count": item["paymentInstruments"][0]["count"],
                                    "insurance_amount": item["paymentInstruments"][0]["amount"]
                                })

                    except Exception:
                        pass

df_insurance = pd.DataFrame(insurance_records)

df_insurance.head()
```

```python
path = "/content/pulse/data/map/user/hover/country/india/state"

map_user_records = []

for state in os.listdir(path):
    state_path = os.path.join(path, state)

    if os.path.isdir(state_path):

        for year in os.listdir(state_path):
            year_path = os.path.join(state_path, year)

            for file in os.listdir(year_path):

                if file.endswith(".json"):

                    quarter = int(file.replace(".json", ""))

                    file_path = os.path.join(year_path, file)

                    with open(file_path, "r") as f:
                        data = json.load(f)

                    try:
                        districts = data["data"]["hoverData"]

                        for district, values in districts.items():
                            map_user_records.append({
                                "state": state.replace("-", " ").title(),
                                "year": int(year),
                                "quarter": quarter,
                                "district": district,
                                "registered_users": values.get("registeredUsers", 0),
                                "app_opens": values.get("appOpens", 0)
                            })

                    except Exception:
                        pass

df_map_user = pd.DataFrame(map_user_records)

df_map_user.head()
```

```python
#1.State-wise transaction volume
#Finding: A small number of states contribute most of the transaction value.
#Business Value: Helps identify high-revenue markets for expansion.

query = """
SELECT state, SUM(Amount) AS total_amount
FROM aggregated_transaction
GROUP BY state;
"""
result = pd.read_sql_query(query, conn)
#result
df=pd.DataFrame(result)
df
```

```python
#2. Quarterly trend analysis
#Finding: Transactions increase in specific quarters (festival/salary periods).
#Business Value: Helps plan campaigns during peak seasons.

query = """
SELECT year, quarter, SUM(amount) AS total_amount
FROM aggregated_transaction
GROUP BY year, quarter
ORDER BY year, quarter;
"""
result = pd.read_sql_query(query, conn)

df1=pd.DataFrame(result)
df1
```

```python
%%writefile app.py
import streamlit as st
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

conn = sqlite3.connect("phonepe.db")

st.title("PhonePe Transaction Insights Dashboard")

menu = st.sidebar.selectbox("Select Analysis", [
    "Home",
    "Transactions Analysis",
    "User Analysis",
    "Insurance Analysis",
    "Geographic Insights",
    "Insights & Findings"
])
# ---------------- HOME ----------------
if menu == "Home":
    st.markdown("""
    ### 📌 Overview
    This dashboard analyzes **PhonePe Pulse data** to understand digital payment trends across India.
    It provides insights into transactions, users, insurance, and geographic performance.
    """)

    # ---------------- KPI METRICS ----------------
    st.subheader("📌 Key Metrics")

    col1, col2, col3, col4 = st.columns(4)

    # Total Transactions
    txn_amount = pd.read_sql_query("SELECT SUM(Amount) AS total FROM aggregated_transaction", conn )["total"][0]
    # Total Users
    users = pd.read_sql_query("SELECT SUM(registered_users) AS total FROM map_user", conn)["total"][0]
    # Insurance Amount
    ins = pd.read_sql_query("SELECT SUM(insurance_amount) AS total FROM aggregated_insurance", conn)["total"][0]
    # Transaction Count
    txn_count = pd.read_sql_query("SELECT SUM(Count) AS total FROM aggregated_transaction", conn)["total"][0]

    col1.metric("💰 Total Transactions", f"{txn_amount:,.0f}")
    col2.metric("👥 Total Users", f"{users:,.0f}")
    col3.metric("🛡️ Insurance Amount", f"{ins:,.0f}")
    col4.metric("📊 Transaction Count", f"{txn_count:,.0f}")

    st.markdown("---")

    # ---------------- OBJECTIVE ----------------
    st.subheader("🎯 Objectives")
    st.markdown("""
    - Analyze transaction growth over years
    - Study user adoption of digital payments
    - Understand insurance usage trends
    - Visualize state-wise performance
    - Identify top performing regions in India
    """)

    st.markdown("---")

    # ---------------- FEATURES ----------------
    st.subheader("⚙️ Dashboard Features")
    st.markdown("""
    ✔ Transaction Analysis (Year, Quarter, State)
    ✔ User Analysis (Brands, Registered Users)
    ✔ Insurance Analysis (Amount & Count)
    ✔ Geographic Insights (India Map Visualization)
    ✔ Top Performers (States & Districts)
    """)

    st.markdown("---")

    # ---------------- FOOTER ----------------
    st.success("📌 Developed for PhonePe Pulse Data Visualization Mini Project")
# ---------------- TRANSACTIONS ----------------
elif menu == "Transactions Analysis":

# ---------------- TRANSACTIONS1 ----------------
    query = """SELECT Year, SUM(Amount) AS total_amount
    FROM aggregated_transaction GROUP BY Year ORDER BY Year;"""

    df2 = pd.read_sql_query(query, conn)

    st.subheader("Year-wise Growth in Transaction Amount")

    fig, ax = plt.subplots(figsize=(10, 5))
    sns.lineplot(data=df2,x="Year",y="total_amount",marker="o",linewidth=3,color="green",ax=ax)
    ax.set_title("Year-wise Growth in Digital Transactions")
    ax.set_xlabel("Year")
    ax.set_ylabel("Total Transaction Amount (₹)")
    plt.tight_layout()
    st.pyplot(fig)
    st.dataframe(df2)

# ---------------- TRANSACTIONS2 ----------------
    df = pd.read_sql_query("SELECT * FROM aggregated_transaction",conn)

    st.subheader("Top 10 States by Transaction Amount")
    top = (df.groupby("State")["Amount"].sum().sort_values(ascending=False).head(10))
    fig, ax = plt.subplots(figsize=(10, 5))
    sns.barplot(x=top.values,y=top.index,palette="Blues_r",ax=ax)
    ax.set_title("Top 10 States by Transaction Amount")
    ax.set_xlabel("Total Transaction Amount")
    ax.set_ylabel("State")
    plt.tight_layout()
    st.pyplot(fig)
    st.dataframe(df)
  # ---------------- TRANSACTIONS3 ----------------
    df_txn = pd.read_sql_query("SELECT * FROM aggregated_transaction", conn)
    st.subheader("Quarterly Transaction Trend")
    trend = df_txn.groupby(["Year","Quarter"])["Amount"].sum().reset_index()
    fig, ax =plt.subplots(figsize=(10,5))
    sns.lineplot(data=trend, x="Quarter", y="Amount", hue="Year")
    plt.title("Quarterly Transaction Trend")
    st.pyplot(fig)
    st.dataframe(trend)

  # ---------------- TRANSACTIONS4 ----------------
    query = """SELECT state, SUM(transaction_amount) AS total_amount FROM top_transaction
    GROUP BY state ORDER BY total_amount desc"""
    df1 = pd.read_sql_query(query, conn)

    st.subheader("State-wise Total Transaction Amount")
    fig, ax =plt.subplots(figsize=(14, 6))
    sns.barplot(data=df1,x='state',y='total_amount',hue='state',palette='viridis',legend=False,ax=ax)
    plt.title('State-wise Total Transaction Amount', fontsize=16)
    plt.xlabel('State', fontsize=12)
    plt.ylabel('Total Transaction Amount', fontsize=12)
    plt.xticks(rotation=90)  # Rotate state names for better readability
    plt.tight_layout()
    st.pyplot(fig)
    st.dataframe(df1)
  # ---------------- USERS ----------------
elif menu == "User Analysis":
  # ---------------- USERS1 ----------------
    query = """SELECT state, SUM(count) AS total_users FROM aggregated_user
    GROUP BY State order by total_users desc"""
    df7 = pd.read_sql_query(query, conn)
    st.subheader("Total Users by State")
    fig, ax =plt.subplots(figsize=(12,6))
    sns.barplot(data=df7,x='State',y='total_users',hue="total_users",palette='viridis')
    plt.title('Total Users by State')
    plt.xlabel('State')
    plt.ylabel('Total Users')
    plt.xticks(rotation=90)  # Rotate state names for better readability
    plt.tight_layout()
    st.pyplot(fig)
    st.dataframe(df7)
  # ---------------- USERS2 ----------------
    df = pd.read_sql_query("SELECT * FROM aggregated_user", conn)
    st.subheader("Device Usage Distribution")
    device = df.groupby("Brand")["Count"].sum().sort_values(ascending=False)
    st.write("Top Device Brands")
    fig, ax = plt.subplots(figsize=(4,4))
    ax.pie(device.head(6),labels=device.head(6).index,autopct="%1.1f%%")
    ax.set_title("Device Usage Distribution")
    st.pyplot(fig)
    st.dataframe(device)
  # ---------------- USERS3 ----------------
    query = """SELECT year, SUM(registered_users) AS total_users
    FROM map_user GROUP BY year;"""
    df4 = pd.read_sql_query(query, conn)
    st.subheader("Year-wise Registered Users")
    fig,ax=plt.subplots(figsize=(10, 5))
    sns.lineplot(data=df4,x='year',y='total_users',marker='o',linewidth=2.5,ax=ax)
    plt.title('Year-wise Registered Users')
    plt.xlabel('Year')
    plt.ylabel('Total Registered Users')
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.tight_layout()
    st.pyplot(fig)
    st.dataframe(df4)
  # ---------------- USERS4 ----------------
    query = """SELECT state, SUM(registered_users) AS total_users
    FROM map_user GROUP BY state order by total_users desc;"""
    df8 = pd.read_sql_query(query, conn)
    st.subheader("Registered Users by State")
    fig, ax =plt.subplots(figsize=(12,6))
    sns.barplot(data=df8,x='state',y='total_users',hue='total_users',palette='crest')
    plt.title('Registered Users by State')
    plt.ylabel('Registered Users')
    plt.xlabel('State')
    plt.xticks(rotation=90)
    plt.tight_layout()
    st.pyplot(fig)
    st.dataframe(df8)
# ---------------- INSURANCE ----------------
elif menu == "Insurance Analysis":

 # ---------------- Insurance 1 ----------------
    query = """SELECT state,SUM(insurance_count) AS total_count FROM aggregated_insurance
    GROUP BY state ORDER BY total_count DESC;"""
    # Extract data from database
    df = pd.read_sql_query(query, conn)
    # Display data
    st.subheader("Insurance Count by State")
    # Line Plot
    fig, ax = plt.subplots(figsize=(12, 6))
    sns.lineplot(data=df,x="state",y="total_count",marker="o",linewidth=2,color="blue",ax=ax)
    ax.set_title("State-wise Insurance Count")
    ax.set_xlabel("State")
    ax.set_ylabel("Total Insurance Count")
    plt.xticks(rotation=90)
    plt.tight_layout()
    st.pyplot(fig)
    st.dataframe(df)
#--------- Insurance 2 ----------------
    st.subheader("Insurance Amount Analysis")
    df_amount = pd.read_sql_query("SELECT * FROM aggregated_insurance",conn)
    top = (df_amount.groupby("state")["insurance_amount"].sum().sort_values(ascending=False).head(10))
    fig, ax = plt.subplots(figsize=(10, 6))
    sns.barplot(x=top.values,y=top.index,palette="viridis",ax=ax)
    ax.set_title("Top 10 States by Insurance Amount")
    ax.set_xlabel("Insurance Amount")
    ax.set_ylabel("State")
    st.pyplot(fig)
    st.dataframe(df_amount)
#--------- Insurance 3 ----------------
    query = """SELECT year, SUM(insurance_amount) AS total_amount
    FROM aggregated_insurance GROUP BY year;"""
    df9 = pd.read_sql_query(query, conn)
    st.subheader("Total Insurance Amount by Year")
    fig, ax = plt.subplots(figsize=(6,4))
    sns.lineplot(data=df9,x='year',y='total_amount',marker='o',linewidth=3,color='royalblue')
    plt.title('Total Insurance Amount by Year', fontsize=14)
    plt.xlabel('Year')
    plt.ylabel('Insurance Amount')
    plt.grid(True, linestyle='--', alpha=0.6)
    st.pyplot(fig)
    st.dataframe(df9)
# ---------------- Geographic Insights ----------------
elif menu == "Geographic Insights":
    st.subheader("Geographic Insights")
    # Aggregate transaction amount by state
    query = """SELECT State,SUM(Amount) AS Total_Amount
    FROM aggregated_transaction GROUP BY State"""
    df6 = pd.read_sql_query(query, conn)
    # State name corrections (if needed)
    df6["State"] = df6["State"].replace({"Andaman & Nicobar Islands": "Andaman and Nicobar Islands",
        "Dadra & Nagar Haveli & Daman & Diu": "Dadra and Nagar Haveli and Daman and Diu"
    })
    st.dataframe(df6)
    # India GeoJSON
    geojson_url = "https://gist.githubusercontent.com/jbrobst/56c13bbbf9d97d187fea01ca62ea5112/raw/e388c4cae20aa53cb5090210a42ebb9b765c0a36/india_states.geojson"

    fig1 = px.choropleth(df6,geojson=geojson_url,featureidkey="properties.ST_NM",locations="State",
        color="Total_Amount",color_continuous_scale="Reds",title="State-wise Transaction Amount")
    fig1.update_geos(fitbounds="locations",visible=False)
    fig1.update_layout(margin={"r":0,"t":50,"l":0,"b":0})
    st.plotly_chart(fig1)

# ---------------- INSIGHTS & FINDINGS ----------------
elif menu == "Insights & Findings":
    st.title("Insights & Findings")
    st.markdown("### 🧠 Key Business Insights")
    st.info("""
    - Digital transactions in India show strong and continuous growth over the years.
    - A few major states contribute most of the total transaction volume.
    - UPI-based payments are driving India's digital economy.
    """)
    st.markdown("---")

    st.markdown("### 📈 Transaction Insights")
    st.info("""
    - Transaction amounts are increasing year by year.
    - Peak usage is often seen in festive quarters (Q3 & Q4).
    - Merchant and peer-to-peer payments dominate usage.
    """)
    st.markdown("---")

    st.markdown("### 🧑 User Insights")
    st.info("""
    - Registered users have grown significantly over time.
    - Android devices dominate mobile usage.
    - Urban states show higher digital adoption than rural areas.
    """)
    st.markdown("---")

    st.markdown("### 🛡️ Insurance Insights")
    st.info("""
    - Insurance transactions are lower compared to payment transactions.
    - Only a few states dominate insurance usage.
    - Insurance adoption is slowly increasing over time.
    """)
    st.markdown("---")

    st.markdown("### 🌍 Geographic Insights")
    st.info("""
    - Western and Southern states lead in digital transactions.
    - Northeastern states show lower adoption rates.
    - Metro cities contribute heavily to total transactions.
    """)
    st.markdown("---")

    st.markdown("### 📊 Growth Insights")
    st.info("""
    - Strong year-on-year growth in transaction volume.
    - Both transaction count and value are increasing steadily.
    - Digital payment adoption is rapidly expanding across India.
    """)

    st.markdown("---")

    st.markdown("### 🎯 Final Conclusion")
    st.success("""
    PhonePe plays a major role in India's digital payment ecosystem.
    The platform is driving financial inclusion, but there is still scope
    for growth in rural regions and insurance adoption.
    """)
```
